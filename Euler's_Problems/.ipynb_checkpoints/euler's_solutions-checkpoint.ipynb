<h2>1.</h2><p>If we List all the natural numbers below 10 that are divisible by 3 or 5, we get 3, 5, 6 and 9.</p>
<p>Find the sum of all the multiplies of 3 and 5 below 1000.</p>

```python
def find_multiplies(x, y):
    start_num = 2
    list_of_multiplies = []
    
    while start_num < 1000:
        if (start_num % x == 0) or (start_num % 5 == 0):
            list_of_multiplies.append(start_num)
        else:
            pass
        start_num+=1
        
    return sum(list_of_multiplies)
```

<h2>2.</h2><p>Fibonacci series are generated by adding the previous two terms.</p>
<p>Find the sum of all even numbered fibbonacci series within 4 million</p>

```python
def fibonacci(p_num):
    num1 = 0
    num2 = 1
    next_num = num2
    count = 1
    fib_series = [0]
    while count < p_num:
        if (next_num % 2 == 0):
            fib_series.append(next_num)
        else:
            pass
        num1,num2 = num2, next_num
        next_num = num1+num2
        count+=1
    return sum(fib_series)
```

<h1>3.</h1><p>If we list the prime factors of 13195, we'll get 5,7,13,29.<p>What is the largerst prime factor of the number 600851475143 ?</p>

```python
def prime_factors(p_factor):
    factors = []
    prime_facs = []
    count = 1
    
    while count < p_factor:
        if p_factor % count == 0:
            factors.append(count)
        count+=2
     
    for fac in factors:
        for j in range(2,fac):
            if fac % j == 0:
                break
        else:
            prime_facs.append(fac)
    
    return max(prime_facs)
```

<h2>4.</h2><p>The largest Palindrome number produced by two digit numbers is 9009.</p><p>Find the Largest Palindrome produced by the product of three digit numbers</p>

```python
def two_digit_Palindrome():
    palindromes = []
    for i in range(100,1000):
        for j in range(100,1000):
            if str(i * j) == str(i * j)[::-1]:
                palindromes.append((i*j))
    return max(palindromes)
```

<h2>5.</h2><p>2520 is the smallest number that can be divided by all the numbers from 1 to 10.</p><p>Find the smallest positive number that can be evenly divisible by all hte numbers from 1 to 20</p>

```python
def even_divisible_nums():
    num = 2
    
    is_found = False
    
    div_lst = []
    
    while not is_found:
        for i in range(1,21):
            if num % i != 0:
                break
        else:
            div_lst.append(num)
                
        if len(div_lst) > 0:
            is_found = True
            return div_lst
        num+=2
```

```python
even_divisible_nums()
```

<h2>6.</h2><p>The sum mof the squares of the 1st 10 natural numbers is 385. The square of the sum of the 1st 10 natural numbers is 3025. Hence the difference between the sums are 2640.</p><p>Find the difference between the sum of the squares of the first one hundred natural numbers and the square of the sum.</p>

```python
import math

def sum_of_squares():    
    sum_of_squares = []
    sqr_of_sums = []
    
    for i in range(0,101):
        sum_of_squares.append(math.pow(i,2))
        sqr_of_sums.append(i)
        
    difference = math.pow(sum(sqr_of_sums),2) - sum(sum_of_squares)
    
    return difference
```
